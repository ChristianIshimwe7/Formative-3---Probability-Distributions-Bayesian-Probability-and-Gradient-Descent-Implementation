```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Data points
x = np.array([1, 3])
y = np.array([3, 6])
n = len(x)

# Initial parameters
m = -1.0
b = 1.0
alpha = 0.1
iterations = 3

# Lists to store results for plotting
m_values = [m]
b_values = [b]
errors = []

# Gradient Descent Function
def gradient_descent(x, y, m, b, alpha, n):
    y_pred = m * x + b
    error = y - y_pred
    mse = np.mean(error ** 2)  # Mean Squared Error
    grad_m = -(2 / n) * np.sum(error * x)  # Gradient for m
    grad_b = -(2 / n) * np.sum(error)      # Gradient for b
    m_new = m - alpha * grad_m
    b_new = b - alpha * grad_b
    return m_new, b_new, mse

# Perform 3 iterations
for i in range(iterations):
    # Unpack the returned values directly
    m_new, b_new, mse = gradient_descent(x, y, m, b, alpha, n)
    # Update m and b with the new values
    m = m_new
    b = b_new
    # Store the updated values
    m_values.append(m)
    b_values.append(b)
    errors.append(mse)
    print(f"Iteration {i+1}: m = {m:.4f}, b = {b:.4f}, MSE = {mse:.4f}")

# Final parameters
print(f"Final parameters after {iterations} iterations: m = {m:.4f}, b = {b:.4f}")

# Visualize Results
plt.figure(figsize=(12, 5))

# Plot 1: Data points and regression line
plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue', label='Data points')
x_range = np.linspace(min(x), max(x), 100)
y_pred_final = m * x_range + b
plt.plot(x_range, y_pred_final, color='red', label=f'Regression line (y = {m:.2f}x + {b:.2f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points and Regression Line')
plt.legend()
plt.grid(True)

# Plot 2: Error over iterations
plt.subplot(1, 2, 2)
plt.plot(range(len(errors) + 1), [0] + errors, marker='o', color='green')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Error Over Iterations')
plt.grid(True)

plt.tight_layout()
plt.show()
```
