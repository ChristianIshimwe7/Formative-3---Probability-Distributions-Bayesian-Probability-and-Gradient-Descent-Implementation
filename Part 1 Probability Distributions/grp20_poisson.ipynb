```python
import matplotlib.pyplot as plt

# Manual implementation of factorial
def factorial(n):
    if n == 0 or n == 1:
        return 1
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

# Manual implementation of e^x using Taylor series approximation (first 10 terms)
def exp(x):
    e = 2.71828  # Approximate value of e
    result = 1.0
    for i in range(1, 10):
        result += (x ** i) / factorial(i)
    return result

# Poisson PMF function
def poisson_pmf(k, lambda_rate):
    return (lambda_rate ** k * exp(-lambda_rate)) / factorial(k)

# Parameters
lambda_rate = 2.0  # Average rate of calls per 10 minutes
k_values = list(range(0, 10))  # Range of occurrences from 0 to 9
probabilities = [poisson_pmf(k, lambda_rate) for k in k_values]

# Answer the question: Probability of exactly 3 calls
p_3_calls = poisson_pmf(3, lambda_rate)
print(f"Probability of exactly 3 calls in 10 minutes: {p_3_calls:.4f}")

# Visualization
plt.figure(figsize=(10, 6))
plt.bar(k_values, probabilities, color='skyblue', edgecolor='black')
plt.xlabel('Number of Calls (k)')
plt.ylabel('Probability P(k)')
plt.title('Poisson Distribution (λ = 2)')
plt.xticks(k_values)
for i, p in enumerate(probabilities):
    plt.text(i, p + 0.01, f'{p:.4f}', ha='center')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
